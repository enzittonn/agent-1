# Test — Executor Node (Step 1.4)

Tests the planner → executor handoff:
1. Planner generates a real plan from a task
2. Executor runs each step one at a time (simulating the graph loop)
3. Verify state transitions at each iteration

```python
import sys
sys.path.insert(0, '../app/backend')

from nodes.planner import planner
from nodes.executor import executor
```

```python
# Initial state — only task is set, everything else is empty
state = {
    "task": "Decision transformer for options trading",
    "messages": [],
    "plan": [],
    "status": "planning",
    "results": [],
    "ui_events": [],
}
```

```python
# --- Planner handoff ---
plan_update = planner(state)
state = {**state, **plan_update}

print(f"Status: {state['status']}")
print(f"Plan ({len(state['plan'])} steps):")
for i, step in enumerate(state['plan'], 1):
    print(f"  {i}. {step}")
```

```python
assert state["status"] == "executing"
assert len(state["plan"]) > 0
print("Planner handoff OK.")
```

```python
# --- Executor loop (simulates graph cycling through supervisor) ---
step_num = 0
while state["status"] == "executing":
    step_num += 1
    current_step = state["plan"][0]
    print(f"\n=== Executing step {step_num}: {current_step} ===")

    exec_update = executor(state)
    state = {**state, **exec_update}

    print(f"Result:\n{state['results'][-1]}")
    print(f"Remaining steps: {len(state['plan'])} | Status: {state['status']}")

print(f"\nDone. Ran {step_num} step(s).")
```

```python
assert state["status"] == "done"
assert state["plan"] == []
assert len(state["results"]) == step_num
print("All assertions passed!")
```

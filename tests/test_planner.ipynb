# Test — Planner Node (Step 1.3)

Directly calls `planner(state)` to verify:
- `plan` is a non-empty `list[str]` of clean step strings
- `status` is `"executing"`
- `messages` contains one `AIMessage`

```python
import sys
sys.path.insert(0, '../app/backend')  # make agent-1/ importable from tests/

from nodes.planner import planner
```

```python
# Minimal state — only the fields planner reads need to be populated
state = {
    "task": "Research how decision transformer can be used for options trading",
    "messages": [],
    "plan": [],
    "status": "planning",
    "results": [],
    "ui_events": [],
}
```

```python
result = planner(state)
result
```

```python
print("Status:", result["status"])
print("Steps:")
for i, step in enumerate(result["plan"], 1):
    print(f"  {i}. {step}")
```

```python
# Assertions
from langchain_core.messages import AIMessage

assert result["status"] == "executing", "status should flip to 'executing'"
assert isinstance(result["plan"], list), "plan should be a list"
assert len(result["plan"]) > 0, "plan should have at least one step"
assert all(isinstance(s, str) and s for s in result["plan"]), "every step should be a non-empty string"
assert len(result["messages"]) == 1, "should append exactly one message"
assert isinstance(result["messages"][0], AIMessage), "message should be AIMessage"

print("All assertions passed!")
```

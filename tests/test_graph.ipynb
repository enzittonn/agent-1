# Graph Inspection

Visualise and inspect the FRIDAY StateGraph — nodes, edges, routing.

```python
import sys
sys.path.insert(0, '../app/backend')

from graph import graph
```

## Diagram

```python
from IPython.display import Image, display

display(Image(graph.get_graph().draw_mermaid_png()))
```

## Nodes

```python
g = graph.get_graph()
print("Nodes:")
for node in g.nodes:
    print(f"  {node}")
```

## Edges

```python
print("Edges:")
for edge in g.edges:
    print(f"  {edge.source} → {edge.target}" + (f"  [condition: {edge.data}]" if edge.data else ""))
```
